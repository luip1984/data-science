```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
sns.set(style="ticks") 
```

Choose one variable and plot that variable four different ways

```python
sns.countplot(y="day", data=tips)
```

```python
sns.catplot(x="day", data=tips)
```

```python
sns.factorplot(x="day" , data=tips, kind = "count");
```

```python
sns.swarmplot(x = "day",  data = tips)
```

Choose two continuous variables, and plot them three different ways.

```python
ax = sns.lineplot(x="tip", y="total_bill", data=tips, palette="tab10", linewidth=2.5)
plt.title('tip amount vs total amount')
ax.set(xlabel='Tip', ylabel='Total')
plt.show()
```

```python
ax = sns.relplot(x="tip", y="total_bill", kind="line", data=tips, ci="sd")
plt.title('tip amount vs total amount')

ax.set(xlabel='Tip', ylabel='Total')
plt.show()
```

```python
ax = sns.scatterplot(x="tip", y="total_bill",data=tips)
plt.title('tip amount vs total amount')
ax.set(xlabel='Tip', ylabel='Total')
plt.show()
```

```python
sns.jointplot(x="tip", y="total_bill", data=tips);
```

Choose one continuous variable and one categorical variable, and plot them six different ways.

```python
sns.catplot(x="day", y="total_bill", data=tips);
```

```python
sns.violinplot(x="day", y="total_bill", data=tips);
```

```python
sns.catplot(x="day", y="total_bill", kind="box", data=tips);
```

```python
g = sns.catplot(x="day", y="total_bill", kind="violin", inner=None, data=tips)
sns.swarmplot(x="day", y="total_bill", color="k", size=3, data=tips, ax=g.ax);
```

```python
sns.catplot(x="day", y="total_bill", kind="bar", data=tips);
```

```python
sns.catplot(x="day", y="total_bill", kind="point", data=tips);
```
